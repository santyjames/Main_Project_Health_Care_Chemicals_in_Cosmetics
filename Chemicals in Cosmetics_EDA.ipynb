# Chemicals in Cosmetics

AIM : To analyze the California Safe Cosmetics Program dataset using Python to identify cosmetic products and brands containing potentially harmful chemical ingredients reported to the California Department of Public Health

```python
import pandas as pd # importing Pandas Libraray
```

```python
df=pd.read_csv(r'C:\Users\User\Desktop\Data Analyst\Final Project\Dataset_Health_Care\Chemicals in Cosmetics\cscpopendata.csv') #Load Dataset
```

```python
df.columns=(df.columns.str.strip()) #used to remove leading and trailing whitespace from the text of all column headers in a DataFrame
```

```python
df=df.replace("�", "", regex=True) # Removed the symbol "�" from entire sheet
```

```python
col1=['ProductName', 'CompanyName','BrandName','PrimaryCategory', 'SubCategory','ChemicalName']

df[col1] = df[col1].apply(lambda x: x.str.title()) # Converts text to Proper Case
```

```python
df.isnull().sum() #Count the number of missing (null/NaN) values in each column
```

```python
df=df.drop(columns=['DiscontinuedDate','ChemicalDateRemoved','CasId', 'CasNumber']) # Delected columns which having more null values 
```

```python
df.columns
```

```python
cols = ['InitialDateReported', 'MostRecentDateReported', 'ChemicalCreatedAt', 'ChemicalUpdatedAt']
df[cols] = df[cols].apply(pd.to_datetime, errors='coerce')  # covert all date column data type in to date time format
```

```python
df.duplicated().sum() #Number of Duplicated rows
```

```python
df=df.drop_duplicates() # Remove duplicated rows
```

```python
df.shape # Dimensions of a DataFrame after removal of duplicate rows
```

```python
df.isnull().sum()
```

```python
df['CSFId']=df['CSFId'].fillna('NA') 
df['CSF']=df['CSF'].fillna('NA') # Replacing missing CSFId and CSF with 'NA' value
```

```python
df['BrandName']=df['BrandName'].fillna(df['BrandName'].mode()[0]) # Replacing missing brand name with mode value
```

```python
df.isnull().sum()
```

```python
df['ChemicalCount'].describe() #Statistical Summary of chemical count
```

```python
#importing library
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
#Univariate Analysis
# To find how the chemical count distribution
plt.hist(df['ChemicalCount'],bins=10,edgecolor='black',color='pink')
plt.title("Distribution of Chemical Count")
plt.xlabel("Chemical Count")
plt.ylabel("Frequency")
plt.figure(figsize=(4,1))
plt.show()
```

```python
#Bivariate Analysis
#Calculate the mean ChemicalCount for all Primary Categories
mean_chemical_count_by_category = df.groupby('PrimaryCategory')['ChemicalCount'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='PrimaryCategory', y='ChemicalCount', data=mean_chemical_count_by_category, hue='PrimaryCategory', palette='viridis', legend=False)
plt.title('Average ChemicalCount by All Primary Categories')
plt.xlabel('Primary Category')
plt.ylabel('Average Chemical Count')
plt.xticks(rotation=90, ha='right') # Rotate labels further for readability
plt.tight_layout()
plt.show()
```

```python
#Top 10 brand name which contain more chemicals
Top10_brand=df.groupby('BrandName')['ChemicalName'].nunique().sort_values(ascending=False).head(10)
Top10_brand_df = Top10_brand.reset_index(name='Number of Chemicals')
fig=px.bar(Top10_brand_df,x='BrandName',y='Number of Chemicals',color='BrandName',title='Top 10 Brand Name by No.of Chemicals')
fig.update_layout(width=900, height=500)
fig.show()
```

```python
#Top 10 brand name which contain less chemicals
lowest10_brand=df.groupby('BrandName')['ChemicalName'].nunique().sort_values(ascending=True).head(10)
lowest10_brand_df = lowest10_brand.reset_index(name='Number of Chemicals')
fig=px.bar(lowest10_brand_df,x='Number of Chemicals',y='BrandName',color='BrandName',title='Top 10 Brand Name Having Less No.of Chemicals')
fig.update_layout(width=900, height=500)
fig.show()
```

```python
# The top 10 brand products that contain the highest number of different chemicals.
Top10_brand_product=df.groupby(['BrandName','ProductName'])['ChemicalName'].nunique().sort_values(ascending=False).head(10)
top10_df = Top10_brand_product.reset_index(name='Number of Chemicals')
fig=px.treemap(top10_df,
                 path=['BrandName','ProductName'],
                 values='Number of Chemicals',
                 title='Top 10 Brands by Number of Chemicals')
fig.update_layout(width=1000, height=600)
fig.show()
```

```python
#Top 5 checmical used in most products
Top5_chemicals=df.groupby('ChemicalName')['ProductName'].nunique().sort_values(ascending=False).head(5)
Top5_chemicals_df=Top5_chemicals.reset_index(name='Number of product') 
fig=px.line(Top5_chemicals_df,x='ChemicalName',y='Number of product',
            title='Top 5 chemicals by products',
            
            markers=True)
fig.update_traces(marker=dict(symbol='star', size=10, color='red')) 
fig.update_traces(line=dict(color='green',width=4, dash='dash'))
fig.update_layout(width=1200,height=700)
```

```python
#Top 5 company which producing more chemical based product
```

```python
top5_company=df.groupby('CompanyName')['ChemicalName'].nunique().sort_values(ascending=False).head(5)
top5_df = top5_company.reset_index(name='Number of Chemicals')
fig = px.pie(top5_df,
             names='CompanyName',
             values='Number of Chemicals',
             title='Top 5 Companies by Number of Chemicals')
fig.update_layout(width=700, height=500)
fig.show()
```

```python
# Top 10 product category and sub category having highest number of chemical used
Top10_product_subcategory=df.groupby(['PrimaryCategory','SubCategory'])['ChemicalName'].nunique().sort_values(ascending=False).head(10)
Top10_product_subcategory_df=Top10_product_subcategory.reset_index(name='Number of Chemicals')
fig=px.scatter_3d(Top10_product_subcategory_df,
                  x='PrimaryCategory',
                  y='SubCategory',
                  z='Number of Chemicals',
                  title="3D Scatter: PrimaryCategory vs SubCategory vs No.of Chemicals",
                  color='PrimaryCategory',
                  size='Number of Chemicals')
fig.update_layout(width=1200,height=700)
```

```python
df1=df.groupby('ChemicalCreatedAt')['ChemicalName'].nunique().sort_values(ascending=False)
df2=df1.reset_index(name='Number of Chemicals')
fig=px.scatter(df2,x='ChemicalCreatedAt',y='Number of Chemicals')
fig.update_layout(width=1000,height=500)
fig.show()
```
